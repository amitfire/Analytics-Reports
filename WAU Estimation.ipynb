```python
def calculate_DAU(df_op):
    users = {}
    for dy in df_op.day.unique():
        users[dy] = [df_op[df_op['day']==dy]['user_id'].nunique(),df_op[df_op['day']==dy]['user_id'],df_op[df_op['day']==dy]['hub_id']]
    dau = pd.DataFrame.from_dict(users,columns=['active_users','users','hub_id'],orient='index')
    dau  =dau.sort_values(by='active_users',ascending = False).reset_index(0).rename(columns={'index':'day'})
    login_count_by_each_admin_most_day = {}
    for i in dau.loc[0]['users']:
        login_count_by_each_admin_most_day[i] = list(dau.loc[0]['users']).count(i)
    
    admin_most = pd.DataFrame(list(login_count_by_each_admin_most_day.keys()),list(login_count_by_each_admin_most_day.values()),columns=['id']).reset_index().rename(columns={'index':'number of logins'})

    admin_most = admin_most.sort_values(by='number of logins',ascending=False)
    hubs = df_op[df_op['user_id'].isin(admin_most.head(10)['id'])]['name'].unique()
    return dau,hubs,admin_most

dau,hubs,active_admins= calculate_DAU(df)

dau 
```

```python
def calculate_WAU(df_op_week):
    users = {}
    for wk in df_op_week.week.unique():
        users[wk] = [df_op_week[df_op_week['week']==wk]['user_id'].nunique(),df_op_week[df_op_week['week']==wk]['user_id'],df_op_week[df_op_week['week']==wk]['hub_id']]
    wau = pd.DataFrame.from_dict(users,columns=['active_users','users','hub_id'],orient='index')
    wau  =wau.sort_values(by='active_users',ascending = False).reset_index(0).rename(columns={'index':'week'})
    login_count_by_each_admin_33_week = {}
    for i in wau.loc[0]['users']:
        login_count_by_each_admin_33_week[i] = list(wau.loc[0]['users']).count(i)

    admin_33 = pd.DataFrame(list(login_count_by_each_admin_33_week.keys()),list(login_count_by_each_admin_33_week.values()),columns=['id']).reset_index().rename(columns={'index':'number of logins'})

    admin_33 = admin_33.sort_values(by='number of logins',ascending=False)
    hubs = df_op_week[df_op_week['user_id'].isin(admin_33.head(10)['id'])]['name'].unique()
    return wau,hubs

wau,hubs2= calculate_WAU(df)

wau 
```
